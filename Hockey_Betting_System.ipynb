```
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Ridge
```

```
path = "https://github.com/Dell0071/hockey/raw/main/hockey.csv"
hockey = pd.read_csv(path)
hockey

```

```
hockey.drop(columns=hockey.columns[5:9])
```

```
hockey['goal_difference'] = hockey['home_goals'] - hockey['visitor_goals']
```

```
hockey.drop(columns=hockey.columns[5:9])
```

```
hockey['home_win'] = np.where(hockey['goal_difference'] > 0, 1, 0)
hockey['home_loss'] = np.where(hockey['goal_difference'] < 0, 1, 0)
```

```
hockey.drop(columns=hockey.columns[5:9])
```

```

hockey_visitor = pd.get_dummies(hockey['visitor'], dtype=np.int64)
hockey_home = pd.get_dummies(hockey['home'], dtype=np.int64)
```

```
hockey_visitor
```

```
hockey_model = hockey_home.sub(hockey_visitor) 
hockey_model['goal_difference'] = hockey['goal_difference']
```

```
hockey_model
```

```

hockey_model.drop('San Jose Sharks',axis=1)
```

```
hockey_train = hockey_model.drop('San Jose Sharks',axis=1)
lr = Ridge(alpha=0.001) 
X = hockey_train.drop(['goal_difference'],axis=1)
y = hockey_model['goal_difference']

lr.fit(X, y)
```

```
hockey_ratings = pd.DataFrame(data={'team': X.columns, 'rating': lr.coef_})
hockey_ratings
```
